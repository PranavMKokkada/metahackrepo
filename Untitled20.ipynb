```python
from unsloth import FastLanguageModel
import torch

# 1. Load the model you just trained
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "code_organism_lora",
    max_seq_length = 2048,
    load_in_4bit = True,
)
FastLanguageModel.for_inference(model)

# 2. Test prompt (heavily corrupted file scenario)
messages = [
    {"role": "system", "content": "You are a self-healing code organism. Detect corruption, repair files, and restore functionality."},
    {
        "role": "user",
        "content": """
Vitality: 18%
Threat Level: CRITICAL

File 'main.py' is severely corrupted:

improt os
improt syss

def calc(x,y)
    prnt("Calculatng...")
    retrn x + / y

if __name__ == "__main__":
    valu = calc(10,)
    print("Result = " + valu)

What is your action?
"""
    },
]

inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors="pt"
).to("cuda")

# 3. Generate the response
outputs = model.generate(
    input_ids=inputs,
    max_new_tokens=180,
    temperature=0.7,
    do_sample=True
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
